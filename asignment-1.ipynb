1. Operations

```python
a=[9,18,27,36,45]
b=[3,6,9,12,15]

print(a+b) #adding or concatinating two lists
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]%b[i])
print(c)  # modulus element wise
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]+b[i])
print(c)  # adding element wise
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]-b[i])
print(c)  # subtracting element wise
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]*b[i])
print(c)  # multiplying element wise
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]//b[i])
print(c)  # floor dividing(without point value) element wise
```

```python
c=2*a
print(c)  # repeating list by multiplying a number
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]/b[i])
print(c)  # dividing element wise
```

```python
c=[]
for i in range(0,len(a)):
  c.append(a[i]/2)
print(c)  # dividing 'a' elements by 2
```

```python
c=[]
for i in range(0,len(a)):
  c.append(b[i]//10)
print(c)  # floor dividing 'b' elements by 10
```

2. Loops

```python
a=[]
for i in range(1,11): #for loop
  a.append(i)
print(a)
```

```python
b=[]
for i in range(10,0,-1): #for loop
  b.append(i)
print(b)
```

```python
a=[4,6,8,10,12,14]
i=0
while(i<len(a)): #while loop
  print(a[i],end=" ")
  i+=1
```

```python
a=[4,6,8,10,12,14]
i=len(a)-1
while(i>-1): #while loop
  print(a[i],end=" ")
  i-=1
```
